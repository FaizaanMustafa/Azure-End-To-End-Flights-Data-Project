```python
dbutils.widgets.text('container','')
dbutils.widgets.text('source_folder','')
dbutils.widgets.text('source_directory','')
dbutils.widgets.text('target_folder','')
dbutils.widgets.text('target_directory','')
container = dbutils.widgets.get('container')
source_folder = dbutils.widgets.get('source_folder')
source_directory = dbutils.widgets.get('source_directory')
target_folder = dbutils.widgets.get('target_folder')
target_directory = dbutils.widgets.get('target_directory')
```

```python
df = spark.readStream.format('cloudFiles') \
        .option('cloudFiles.format','csv') \
        .option('cloudFiles.schemaLocation',f'abfss://{container}@adls7428.dfs.core.windows.net/{source_folder}/checkpoint_{source_directory}') \
        .option('cloudFiles.schemaEvolutionMode','rescue') \
        .load(f'abfss://{container}@adls7428.dfs.core.windows.net/{source_folder}/{source_directory}')
df.display()
```

```python
df.writeStream.format('delta').outputMode('append').option('checkpointLocation',f'abfss://{container}@adls7428.dfs.core.windows.net/{target_folder}/checkpoint_{target_directory}').option('path',f'abfss://{container}@adls7428.dfs.core.windows.net/{target_folder}/{target_directory}').trigger(availableNow=True).toTable(f'flights.bronze.source_{target_directory}')
```
